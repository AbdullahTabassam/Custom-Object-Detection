## Losses

The losses can be observable from the tensorboard above. For convenience, screenshots of all the plots have been saved in a folder and can be displayed to be observed.

```python
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
%matplotlib inline

plt.figure(figsize=(20,10))
plt.suptitle('Losses', fontsize=25)

ttl_loss = mpimg.imread("tensorboard_plots/SSD/Loss/total_loss.PNG")
clf_loss = mpimg.imread("tensorboard_plots/SSD/Loss/classification_loss.PNG")
reg_loss = mpimg.imread("tensorboard_plots/SSD/Loss/regularization_loss.PNG")
loc_loss = mpimg.imread("tensorboard_plots/SSD/Loss/localization_loss.PNG")
plt.axis('off')


ax1 = plt.subplot(221)
ax1.imshow(clf_loss)

ax2 = plt.subplot(222)
ax2.imshow(loc_loss)
 
ax3 = plt.subplot(223)
ax3.imshow(reg_loss)

ax4 = plt.subplot(224)
ax4.imshow(ttl_loss)
```

The model is a good fit as the validation loss curve (blue) remains parallel and slightly ablove the training loss curve (orange) and both start at a high loss value but the loss is reduced over time and comes to a plateu after some time which is a sign for a good fit model.

```python
plt.figure(figsize=(9,9))
plt.title('Learning Rate', fontsize=12)
plt.axis('off')

LR = mpimg.imread("tensorboard_plots/SSD/LR/learning_rate.PNG")
plt.imshow(LR)
```

## Evaluation

```python
plt.figure(figsize=(20,10))
plt.suptitle('mAp', fontsize=25)
plt.axis('off')

mAp   = mpimg.imread("tensorboard_plots/SSD/mAP/mAp.PNG")
mAp_1 = mpimg.imread("tensorboard_plots/SSD/mAP/mAp_large.PNG")
mAp_2 = mpimg.imread("tensorboard_plots/SSD/mAP/mAp_medium.PNG")
mAp_3 = mpimg.imread("tensorboard_plots/SSD/mAP/mAp_small.PNG")
plt.axis('off')


ax1 = plt.subplot(221)
ax1.imshow(mAp)

ax2 = plt.subplot(222)
ax2.imshow(mAp_1)
 
ax3 = plt.subplot(223)
ax3.imshow(mAp_2)

ax4 = plt.subplot(224)
ax4.imshow(mAp_3)
```

```python
plt.figure(figsize=(20,10))
plt.suptitle('mAp (IOU)', fontsize=25)
plt.axis('off')

mAp_70_IOU   = mpimg.imread("tensorboard_plots/SSD/mAP/mAp_70_IOU.PNG")
mAp_50_IOU   = mpimg.imread("tensorboard_plots/SSD/mAP/mAp_50_IOU.PNG")

plt.axis('off')


ax1 = plt.subplot(221)
ax1.imshow(mAp_70_IOU )

ax2 = plt.subplot(222)
ax2.imshow(mAp_50_IOU )
```

The mAP values are quite good for this model. mAP values are 90.2% mAp @50IOU and 68.4% mAp @75IOU. A 90% mAP means that out of every 100 bird predictions 90 were correct and 10 were incorrect.
